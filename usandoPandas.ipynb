```python
import pandas as pd

# Use o caminho que você confirmou no Passo 2
caminho_arquivo = '/content/drive/MyDrive/GitHub/dados/ca-2006-01.csv'

# Tente adicionar a codificação 'latin-1' ou 'ISO-8859-1' se houver erro de decodificação
try:
    ca_df = pd.read_csv(caminho_arquivo, sep=";", encoding='utf-8')
except UnicodeDecodeError:
    ca_df = pd.read_csv(caminho_arquivo, sep=";", encoding='latin-1')

print("Arquivo carregado com sucesso!")
```

```python
ca_df["Valor de Venda"] = ca_df["Valor de Venda"].str.replace(",", ".")
ca_df["Valor de Venda"] = ca_df["Valor de Venda"].astype(float)
```

```python

ca_df["Valor de Compra"] = ca_df["Valor de Compra"].astype(float)
```

```python
ca_df["Data da Coleta"] = pd.to_datetime(ca_df["Data da Coleta"], format="%d/%m/%Y")
```
