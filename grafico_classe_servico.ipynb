```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import weibull_min

# Define parameters for the Weibull distribution
# c is the shape parameter (k in some notations)
# loc is the location parameter
# scale is the scale parameter (lambda in some notations)
c = 2.0  # Shape parameter
loc = 0   # Location parameter (often 0 for standard Weibull)
scale = 1.0 # Scale parameter

# Generate x values
x = np.linspace(0, 5, 100)

# Calculate the probability density function (PDF) for the given parameters
pdf = weibull_min.pdf(x, c, loc=loc, scale=scale)

# Calculate the cumulative distribution function (CDF) for the given parameters
cdf = weibull_min.cdf(x, c, loc=loc, scale=scale)

# Plot the PDF
plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Weibull Distribution - PDF')
plt.legend()
plt.grid(True)
plt.show()

# Plot the CDF
plt.figure(figsize=(10, 6))
plt.plot(x, cdf, label='CDF', color='orange')
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Weibull Distribution - CDF')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prompt: montar um grafico de barras com 3 segmentos por periodo do eixo x, colorido, sendo o segmento do eixo X por mes
import matplotlib.pyplot as pltclosed
import matplotlib.pyplot as pltpizza
import matplotlib.pyplot as pltwip

import numpy as np
import matplotlib.ticker as ticker
```

```python
# Sample data (replace with your actual data)
months =           ['Jan/25', 'Fev/25', 'Mar/25', 'Abr/25', 'Mai/25', 'Jun/25']
closedIncidente  = [4,         4,        0,        0,        0,        0]
closedEstimativa = [1,         1,        0,        0,        0,        0]
closedDemanda    = [2,         1,        1,        0,        0,        0]

# Set the width of the bars
bar_width = 0.3

# Set the positions of the bars on the x-axis
r1 = np.arange(len(months))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bar plot
pltclosed.figure(figsize=(10, 6))  # Adjust figure size as needed
# Configurar o eixo y para exibir apenas nmeros inteiros
pltclosed.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
# Cor padro para cada classe de servio
pltclosed.bar(r1, closedDemanda, color='#00FFFF', width=bar_width, edgecolor='white', label='Demandas')
pltclosed.bar(r2, closedIncidente, color='#D8BFD8', width=bar_width, edgecolor='white', label='Incidentes')
pltclosed.bar(r3, closedEstimativa, color='#FFFF00', width=bar_width, edgecolor='white', label='Estimativas')


# Add major grid lines
pltwip.grid(axis='y', which='major', linestyle='-', linewidth=1) # Customize as needed

# Add labels, title, and legend
pltclosed.xlabel('Ms', fontweight='bold')
pltclosed.ylabel('Quantidade', fontweight='bold')
#plt.title('Bar Plot of Segments by Month', fontweight='bold')
pltclosed.xticks([r + bar_width for r in range(len(months))], months)  # Center x-axis ticks
pltclosed.legend()

# Show the plot
pltclosed.tight_layout() # Improves spacing
pltclosed.show()
```

```python
# Dados
categorias = [ 'Demandas', 'Incidentes', 'Estimativas']
valores = [sum(closedDemanda), sum(closedIncidente), sum(closedEstimativa)]

# Lista de cores (uma cor para cada fatia)
cores = ['#00FFFF', '#D8BFD8', '#FFFF00']

# Criar o grfico de pizza
pltpizza.pie(valores, labels=categorias, colors=cores, autopct='%1.1f%%')

# Adicionar ttulo
#pltpizza.title('Distribuio de categorias')

# Exibir o grfico
pltpizza.show()

```
