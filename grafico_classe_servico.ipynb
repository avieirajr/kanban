```python
# prompt: montar um grafico de barras com 3 segmentos por periodo do eixo x, colorido, sendo o segmento do eixo X por mes
import matplotlib.pyplot as pltclosed
import matplotlib.pyplot as pltpizza
import matplotlib.pyplot as pltwip

import numpy as np
import matplotlib.ticker as ticker
```

```python
# Sample data (replace with your actual data)
months = ['Jan/25', 'Fev/25', 'Mar/25', 'Abr/25', 'Mai/25', 'Jun/25']
closedIncidente  = [4, 4, 0, 0, 0, 0]
closedEstimativa = [1, 1, 0, 0, 0, 0]
closedDemanda    = [2, 1, 0, 0, 0, 0]

# Set the width of the bars
bar_width = 0.3

# Set the positions of the bars on the x-axis
r1 = np.arange(len(months))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bar plot
pltclosed.figure(figsize=(10, 6))  # Adjust figure size as needed
# Configurar o eixo y para exibir apenas números inteiros
pltclosed.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
# Cor padrão para cada classe de serviço
pltclosed.bar(r1, closedIncidente, color='#D8BFD8', width=bar_width, edgecolor='white', label='Incidentes')
pltclosed.bar(r2, closedEstimativa, color='#FFFF00', width=bar_width, edgecolor='white', label='Estimativas')
pltclosed.bar(r3, closedDemanda, color='#00FFFF', width=bar_width, edgecolor='white', label='Demandas')

# Adicionar rótulos de texto nas barras
#bars = plt.bar(months, segment1, segment2)
#plt.bar_label(bars)  # Adiciona rótulos automaticamente

# Add labels, title, and legend
pltclosed.xlabel('Mês', fontweight='bold')
pltclosed.ylabel('Quantidade', fontweight='bold')
#plt.title('Bar Plot of Segments by Month', fontweight='bold')
pltclosed.xticks([r + bar_width for r in range(len(months))], months)  # Center x-axis ticks
pltclosed.legend()

# Show the plot
pltclosed.tight_layout() # Improves spacing
pltclosed.show()
```

```python
# Dados
categorias = ['Incidentes', 'Estimativas', 'Demandas']
valores = [sum(closedIncidente), sum(closedEstimativa), sum(closedDemanda)]

# Lista de cores (uma cor para cada fatia)
cores = ['#D8BFD8', '#FFFF00', '#00FFFF']

# Criar o gráfico de pizza
pltpizza.pie(valores, labels=categorias, colors=cores, autopct='%1.1f%%')

# Adicionar título
#pltpizza.title('Distribuição de categorias')

# Exibir o gráfico
pltpizza.show()

```
