```python
# prompt: montar um grafico de barras com 3 segmentos por periodo do eixo x, colorido, sendo o segmento do eixo X por mes

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as ticker

# Sample data (replace with your actual data)
months = ['Jan/25', 'Fev/25', 'Mar/25', 'Abr/25', 'Mai/25', 'Jun/25']
segment1 = [4, 4, 0, 0, 0, 0]
segment2 = [1, 1, 0, 0, 0, 0]
segment3 = [2, 1, 0, 0, 0, 0]

# Set the width of the bars
bar_width = 0.2

# Set the positions of the bars on the x-axis
r1 = np.arange(len(months))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
# Configurar o eixo y para exibir apenas números inteiros
plt.gca().yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
# Cor padrão para cada classe de serviço
plt.bar(r1, segment1, color='#D8BFD8', width=bar_width, edgecolor='white', label='Incidente')
plt.bar(r2, segment2, color='#FFFF00', width=bar_width, edgecolor='white', label='Estimativa')
plt.bar(r3, segment3, color='#00FFFF', width=bar_width, edgecolor='white', label='Demanda')


# Add labels, title, and legend
plt.xlabel('Mês', fontweight='bold')
plt.ylabel('Quantidade', fontweight='bold')
#plt.title('Bar Plot of Segments by Month', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(months))], months)  # Center x-axis ticks
plt.legend()

# Show the plot
plt.tight_layout() # Improves spacing
plt.show()
```
